```python
condition1= data['sup_dist'] == 'District 1'
value_counts_1 = data[condition1]['ami_category'].value_counts()
print(value_counts_1)

condition2= data['sup_dist'] == 'District 2'
value_counts_2 = data[condition1]['ami_category'].value_counts()
print(value_counts_2)

condition3= data['sup_dist'] == 'District 3'
value_counts_3 = data[condition3]['ami_category'].value_counts()
print(value_counts_3)

condition4= data['sup_dist'] == 'District 4'
value_counts_4 = data[condition1]['ami_category'].value_counts()
print(value_counts_4)
```

the output is how many tracts in district 1/2/3/4 are at low income/ very low income/ above moderate/ extremely low income and moderate income

```python
import matplotlib.pyplot as plt

value_counts_df = pd.DataFrame({
    'district 1': data[condition1]['ami_category'].value_counts(),
    'district 2': data[condition2]['ami_category'].value_counts(),
    'district 3': data[condition3]['ami_category'].value_counts(),
    'district 4': data[condition4]['ami_category'].value_counts(),
})
value_counts_df.plot(kind='bar')

plt.xlabel('catogory')
plt.ylabel('number')

plt.show()
```

This bar chart shows how median income distribute among different districts.
